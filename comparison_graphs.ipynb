```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['figure.dpi'] = 150
```

```python
#Control Group
No_Campagin = "Env_Aw_English experiment3-None_by_sean-table.csv"
All = "Env_Aw_English experiment3-All_by_sean-table.csv"
High = "Env_Aw_English experiment3-High_by_sean-table.csv"
Medium = "Env_Aw_English experiment3-Medium_by_sean-table.csv"
Low = "Env_Aw_English experiment3-Low_by_sean-table.csv"
```

```python
#Experimental Group
No_Campagin_E = "Env_Aw_English experiment4-7days-None_by_sean-table.csv"
All_E = "Env_Aw_English experiment4-7days-All_by_sean-table.csv"
High_E = "Env_Aw_English experiment4-7days-High_by_sean-table.csv"
Medium_E = "Env_Aw_English experiment4-7days-Medium_by_sean-table.csv"
Low_E = "Env_Aw_English experiment4-7days-Low_by_sean-table.csv"
```

```python
data_baseline = pd.read_csv(No_Campagin, header=6)
data_all = pd.read_csv(All, header=6)
data_high = pd.read_csv(High, header=6)
data_medium = pd.read_csv(Medium, header=6)
data_low = pd.read_csv(Low, header=6)
```

```python
data_baseline_E = pd.read_csv(No_Campagin_E, header=6)
data_all_E = pd.read_csv(All_E, header=6)
data_high_E = pd.read_csv(High_E, header=6)
data_medium_E = pd.read_csv(Medium_E, header=6)
data_low_E = pd.read_csv(Low_E, header=6)
```

```python
def get_eco_level_pop(data, column):
  result = data.groupby('[step]')[column].mean()
  return result
```

```python
data_baseline_eco_level_E = get_eco_level_pop(data_baseline,column="eco_level_pop")
data_all_eco_level_E = get_eco_level_pop(data_all,column="eco_level_pop")
data_high_eco_level_E = get_eco_level_pop(data_high,column="eco_level_pop")
data_medium_eco_level_E = get_eco_level_pop(data_medium,column="eco_level_pop")
data_low_eco_level_E = get_eco_level_pop(data_low,column="eco_level_pop")
df1 = pd.DataFrame({'baseline_eco_level': data_baseline_eco_level_E})
df2 = pd.DataFrame({'all_eco_level': data_all_eco_level_E})
df3 = pd.DataFrame({'high_eco_level': data_high_eco_level_E})
df4 = pd.DataFrame({'medium_eco_level': data_medium_eco_level_E})
df5 = pd.DataFrame({'low_eco_level': data_low_eco_level_E})
control = pd.concat([df1, df2, df3, df4, df5], axis= 1)
```

```python
data_baseline_eco_level_E = get_eco_level_pop(data_baseline_E,column="eco_level_pop")
data_all_eco_level_E = get_eco_level_pop(data_all_E,column="eco_level_pop")
data_high_eco_level_E = get_eco_level_pop(data_high_E,column="eco_level_pop")
data_medium_eco_level_E = get_eco_level_pop(data_medium_E,column="eco_level_pop")
data_low_eco_level_E = get_eco_level_pop(data_low_E,column="eco_level_pop")
df1 = pd.DataFrame({'baseline_eco_level': data_baseline_eco_level_E})
df2 = pd.DataFrame({'all_eco_level': data_all_eco_level_E})
df3 = pd.DataFrame({'high_eco_level': data_high_eco_level_E})
df4 = pd.DataFrame({'medium_eco_level': data_medium_eco_level_E})
df5 = pd.DataFrame({'low_eco_level': data_low_eco_level_E})
experiment = pd.concat([df1, df2, df3, df4, df5], axis= 1)
```

```python
# Plot the data with labels
plt.figure(figsize=(10, 6))
plt.plot(control['baseline_eco_level'], label='No Campaign Eco Level - Monthly', color='blue')
plt.plot(control['all_eco_level'], label='All Eco Level Campaign - Monthly', color='orange')
plt.plot(control['high_eco_level'], label='High Eco Level Campaign - Monthly', color='green')
plt.plot(control['medium_eco_level'], label='Medium Eco Level Campaign - Monthly', color = 'red')
plt.plot(control['low_eco_level'], label='Low Eco Level Campaign - Monthly', color='purple')

plt.plot(experiment['baseline_eco_level'], label='No Campaign Eco Level - Weekly', color='blue', linestyle='--')
plt.plot(experiment['all_eco_level'], label='All Eco Level Campaign - Weekly', color='orange', linestyle='--')
plt.plot(experiment['high_eco_level'], label='High Eco Level Campaign- Weekly', color='green', linestyle='--')
plt.plot(experiment['medium_eco_level'], label='Medium Eco Level Campaign- Weekly', color='red', linestyle='--')
plt.plot(experiment['low_eco_level'], label='Low Eco Level Campaign- Weekly', color='purple', linestyle='--')
# Add a legend
plt.legend()

# Set the title and labels
plt.title('Eco Level Comparison - Weekly VS Monthly')
plt.xlabel('Days/Ticks')
plt.ylabel('Average eco-level of total population')
plt.grid(True)
plt.show()
```
