```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['figure.dpi'] = 150
```

```python
data_baseline = pd.read_csv("Env_Aw_English Baseline-table.csv", header=6)
data_all = pd.read_csv("Env_Aw_English experiment-All-table.csv", header=6)
data_high = pd.read_csv("Env_Aw_English experiment-High-table.csv", header=6)
data_medium = pd.read_csv("Env_Aw_English experiment-Medium-table.csv", header=6)
data_low = pd.read_csv("Env_Aw_English experiment-Low-table.csv", header=6)
```

```python

def get_eco_level_pop(data, column):
  return data[column]
```

```python
data_baseline_eco_level = get_eco_level_pop(data_baseline,column="eco_level_pop")
data_all_eco_level = get_eco_level_pop(data_all,column="eco_level_pop")
data_high_eco_level = get_eco_level_pop(data_high,column="eco_level_pop")
data_medium_eco_level = get_eco_level_pop(data_medium,column="eco_level_pop")
data_low_eco_level = get_eco_level_pop(data_low,column="eco_level_pop")
```

```python
df1 = pd.DataFrame({'baseline_eco_level': data_baseline_eco_level})
df2 = pd.DataFrame({'all_eco_level': data_all_eco_level})
df3 = pd.DataFrame({'high_eco_level': data_high_eco_level})
df4 = pd.DataFrame({'medium_eco_level': data_medium_eco_level})
df5 = pd.DataFrame({'low_eco_level': data_low_eco_level})
experiment = pd.concat([df1, df2, df3, df4, df5], axis= 1)
experiment
```

```python
# Plot the data with labels
plt.figure(figsize=(10, 6))
plt.plot(experiment['baseline_eco_level'], label='Baseline Eco Level')
plt.plot(experiment['all_eco_level'], label='All Eco Level')
plt.plot(experiment['high_eco_level'], label='High Eco Level')
plt.plot(experiment['medium_eco_level'], label='Medium Eco Level')
plt.plot(experiment['low_eco_level'], label='Low Eco Level')

# Add a legend
plt.legend()

# Set the title and labels
plt.title('Eco Level Comparison')
plt.xlabel('Index')
plt.ylabel('Eco Level')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with 5 subplots (one for each graph)
fig, axs = plt.subplots(5, 1, figsize=(10, 12), sharex=True)

# Define a list of colors for each graph
colors = ['blue', 'orange', 'green', 'red', 'purple']

# Plot the data for each eco level in a separate subplot with different colors
for i, (label, color) in enumerate(zip(['baseline_eco_level', 'all_eco_level', 'high_eco_level', 'medium_eco_level', 'low_eco_level'], colors)):
    axs[i].plot(experiment[label], label=label, color=color)
    axs[i].set_ylabel(label)
    axs[i].grid(True)

# Set the title and labels for the entire figure
fig.suptitle('Eco Level Comparison')
axs[4].set_xlabel('Index')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```
